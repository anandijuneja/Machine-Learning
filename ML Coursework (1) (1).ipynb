```
receipt_total = spark.sql("""
SELECT receipt_id, SUM(qty) as total_quantity, SUM(value) as total_value, COUNT(product_code) as product_variety
FROM receipt_lines
GROUP BY receipt_id
ORDER BY receipt_id""").toPandas()
receipt_total
```

```
current_date = max(receipts['purchased_at'])
current_date
```

```
import pandas as pd
receipts['day'] = ((current_date - receipts['purchased_at'])/ pd.Timedelta(days=1)).astype(int)
receipts
```

```
new = pd.merge(receipt_total, receipts[['receipt_id', 'purchased_at', 'customer_id', 'day']], on = ['receipt_id'], how = 'inner')
new = new.sort_values('purchased_at', ascending = False)
new.dtypes
```

```
new
```

```
active_c = new[(new['day'] >= 0) & (new['day'] <= 16)]['customer_id'].unique()
active_receipts = new[new['customer_id'].isin(active_c)]
active_receipts
```

```
def get_dataset(ref_date, tws, ows):
    # Group the data by customer_id
    grouped_df = active_receipts.groupby('customer_id')
    output_spending = grouped_df.apply(lambda x: x[(x['day'] < ref_date) & (x['day'] >= (ref_date - ows))]\
                                       ['total_value'].sum()).reset_index(name = 'output_spending')
    output_quantity = grouped_df.apply(lambda x: x[(x['day'] < ref_date) & (x['day'] >= (ref_date - ows))]\
                                       ['total_quantity'].sum()).reset_index(name = 'output_quantity')
    output_frequency = grouped_df.apply(lambda x: x[(x['day'] < ref_date) & (x['day'] >= (ref_date - ows))]\
                                        ['day'].nunique()).reset_index(name = 'output_frequency')
    output_churn = grouped_df.apply(lambda x: 1 if x[(x['day'] < ref_date) & (x['day'] >= (ref_date - ows))]\
                                    ['day'].nunique() == 0 else 0).reset_index(name = 'output_churn')
    f1_spending = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + tws)) & (x['day'] >= ref_date)]\
                                       ['total_value'].sum()).reset_index(name = 'f1_spending')
    f1_quantity = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + tws)) & (x['day'] >= ref_date)]\
                                       ['total_quantity'].sum()).reset_index(name='f1_quantity')
    f1_frequency = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + tws)) & (x['day'] >= ref_date)]\
                                        ['day'].nunique()).reset_index(name='f1_frequency')
    f1_churn = grouped_df.apply(lambda x: 1 if x[(x['day'] < (ref_date + tws)) & (x['day'] >= ref_date)]\
                                    ['day'].nunique() == 0 else 0).reset_index(name='f1_churn')
    f2_spending = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + 2 * tws)) & (x['day'] >= (ref_date + tws))]\
                                       ['total_value'].sum()).reset_index(name='f2_spending')
    f2_quantity = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + 2 * tws)) & (x['day'] >= (ref_date + tws))]\
                                       ['total_quantity'].sum()).reset_index(name='f2_quantity')
    f2_frequency = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + 2 * tws)) & (x['day'] >= (ref_date + tws))]\
                                        ['day'].nunique()).reset_index(name='f2_frequency')
    f2_churn = grouped_df.apply(lambda x: 1 if x[(x['day'] < (ref_date + 2 * tws)) & (x['day'] >= (ref_date + tws))]\
                                    ['day'].nunique() == 0 else 0).reset_index(name='f2_churn')
    f3_spending = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + 3 * tws)) & (x['day'] >= (ref_date + 2 * tws))]\
                                       ['total_value'].sum()).reset_index(name='f3_spending')
    f3_quantity = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + 3 * tws)) & (x['day'] >= (ref_date + 2 * tws))]\
                                       ['total_quantity'].sum()).reset_index(name='f3_quantity')
    f3_frequency = grouped_df.apply(lambda x: x[(x['day'] < (ref_date + 3 * tws)) & (x['day'] >= (ref_date + 2 * tws))]\
                                        ['day'].nunique()).reset_index(name='f3_frequency')
    f3_churn = grouped_df.apply(lambda x: 1 if x[(x['day'] < (ref_date + 3 * tws)) & (x['day'] >= (ref_date + 2 * tws))]\
                                    ['day'].nunique() == 0 else 0).reset_index(name='f3_churn')
    
    output_df = pd.merge(output_spending, output_quantity, on='customer_id')
    output_df = pd.merge(output_df, output_frequency, on='customer_id')
    output_df = pd.merge(output_df, output_churn, on='customer_id')
    output_df = pd.merge(output_df, f1_spending, on='customer_id')
    output_df = pd.merge(output_df, f1_quantity, on='customer_id')
    output_df = pd.merge(output_df, f1_frequency, on='customer_id')
    output_df = pd.merge(output_df, f1_churn, on='customer_id')
    output_df = pd.merge(output_df, f2_spending, on='customer_id')
    output_df = pd.merge(output_df, f2_quantity, on='customer_id')
    output_df = pd.merge(output_df, f2_frequency, on='customer_id')
    output_df = pd.merge(output_df, f2_churn, on='customer_id')
    output_df = pd.merge(output_df, f3_spending, on='customer_id')
    output_df = pd.merge(output_df, f3_quantity, on='customer_id')
    output_df = pd.merge(output_df, f3_frequency, on='customer_id')
    output_df = pd.merge(output_df, f3_churn, on='customer_id')
    
    return output_df.drop(columns = ['output_spending', 'customer_id', 'output_quantity', 'output_frequency', 'output_churn'], inplace = False), output_df['output_churn']
```

```
# Applying Random Forest Algorithm

output_window_size = 16
tumbling_window_size = 30
now = 0
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import numpy as np
from sklearn.metrics import classification_report

from sklearn.model_selection import train_test_split
model = RandomForestClassifier()
test = get_dataset( now + output_window_size, tumbling_window_size, output_window_size )
train = get_dataset( now + 2 * output_window_size, tumbling_window_size, output_window_size )
model.fit( train[0], train[1] )

predict = model.predict(test[0])
accuracy = metrics.accuracy_score(test[1],predict)
print(classification_report(test[1], predict))
print("accuracy", np.mean(accuracy))
```

```
from inspect import classify_class_attrs
from sklearn.svm import SVC

clf = SVC()
clf.fit(train[0], train[1])
clf_predict = clf.predict(test[0])
accuracy_svc =metrics.accuracy_score(test[1], clf_predict)
print(classification_report(test[1], clf_predict))
print('SVC accuracy', np.mean(accuracy_svc))
```

```
from sklearn.model_selection import GridSearchCV
import numpy as np

max_features_range = np.arange(1,6,1)
n_estimators_range = np.arange(10,210,10)
param_grid = dict(max_features=max_features_range, n_estimators=n_estimators_range)

rf = RandomForestClassifier()

grid = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)

grid.fit(train[0], train[1])

print("The best parameters are %s with a score of %0.2f"
      % (grid.best_params_, grid.best_score_))
```

```
grid_results = pd.concat([pd.DataFrame(grid.cv_results_["params"]),pd.DataFrame(grid.cv_results_["mean_test_score"], columns=["Accuracy"])],axis=1)
grid_results
```
